we are trying to draw a rectangle across a face

```python
faceCascade =cv2.CascadeClassifier(cv2.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
faces = faceCascade.detectMultiScale(gray,1.3,5)
# draw a rectangle across the face
for(x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
# use putText for inserting text on video
cv2.putText(img, predictions['dominant_emotion'], (0,50), font, 1, (0, 0, 255), 2, cv2.LINE_4);
```
